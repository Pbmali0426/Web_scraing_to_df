```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
web_page=requests.get('https://etenders.gov.in/eprocure/app').text
soup=BeautifulSoup(web_page,'lxml')
data=[]
for i in range(len(soup.find_all(attrs={'class':'even'}))):
    text=soup.find_all(attrs={'class':'even'})[i].text
    data.append(text)
```

```python
data
Tender_title=[]
Reference_no=[]
Closing_Date=[]
Bid_Opening_Date=[]

for i in range(len(data)):
    s=data[i].split('\n')[1:-1]
    Tender_title.append(s[0])
    Reference_no.append(s[1])
    Closing_Date.append(s[2])
    Bid_Opening_Date.append(s[3])
    
Dict_data={'Tender_title':Tender_title,'Reference_no':Reference_no,'Closing_Date':Closing_Date,'Bid_Opening_Date':Bid_Opening_Date}
df=pd.DataFrame(Dict_data)
df
```
